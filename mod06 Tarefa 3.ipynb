```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib widget
```

```python
df_sinasc = pd.read_csv("b1abb33c4f.csv", sep=',')
df_sinasc.head()
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
df_sinasc.loc[(df_sinasc['APGAR5'] >= 0) & (df_sinasc['APGAR5'] <= 3), 'CATASFIXIA'] = 'asfixia severa'
df_sinasc.loc[(df_sinasc['APGAR5'] >= 4) & (df_sinasc['APGAR5'] <= 5), 'CATASFIXIA'] = 'asfixia moderada'
df_sinasc.loc[(df_sinasc['APGAR5'] >= 6) & (df_sinasc['APGAR5'] <= 7), 'CATASFIXIA'] = 'asfixia leve'
df_sinasc.loc[(df_sinasc['APGAR5'] >= 8) & (df_sinasc['APGAR5'] <= 10), 'CATASFIXIA'] = 'normal'
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
sns.boxplot(data=df_sinasc, x='GESTACAO', y='APGAR5', hue='SEXO')

plt.xlabel('Tempo de Gestação')
plt.xticks(rotation=10, fontsize=5)
plt.ylabel('APGAR5')
plt.title('Distribuição de APGAR5 por Sexo e Tempo de Gestação')
plt.show()
```

```python
plt.close()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab = pd.crosstab(df_sinasc['GESTACAO'], df_sinasc['APGAR5'])
tab
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
perc = tab.div(tab.sum(axis= 1), axis=0)
perc
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Ordenando o eixo X com a lista das categorias de tempo de gestação.
perc.plot.bar()
plt.xticks(rotation= 5, fontsize=5)
```

Como podemos analisar no gráfico acima abaixo de 37 semanas vemos que a um grande percentual de APGAR cada vez menores indicando maiores indice de asfixia, entretanto no intervalo entree 32 a 36 semanas podemos observar que possui um alto nivel de concentração dos bebes como normal, ou seja, é uma afirmação correto apesar de que conforme vai se aproximando do periodo normal de parto há menos complicações no nascimento do mesmo, antes disso apresenta um elevado indice de concentração de partos pré maturos.
