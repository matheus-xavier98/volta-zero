```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_dados = pd.read_csv("b1abb33c4f.csv", sep= ',')
base_dados.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
#Como estamos lindando com uma grande quantidade de dados antes de plotar o grafico estarei fazendo a limpeza dos dados.


colunas_desejadas = ["APGAR1", "APGAR5", "SEXO", "GESTACAO", "ESTCIVMAE"]

base_dados.dropna(subset = colunas_desejadas, inplace= True)

#Devido ao problema referente ao tamanho dos graficos irei criar um gráfico para cada Váriavel.
```

```python
sns.countplot(data=base_dados, x='APGAR1')
plt.title('APGAR1')
plt.show()
```

```python
sns.countplot(data=base_dados, x='APGAR5')
plt.title('APGAR5')
plt.show()
```

```python
sns.countplot(data=base_dados, x='SEXO')
plt.title('SEXO')
plt.show()
```

```python
sns.countplot(data=base_dados, x='GESTACAO')
plt.title('GESTACAO')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=base_dados, x='ESTCIVMAE')
plt.title('ESTCIVMAE')
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(data=base_dados, x='APGAR5', y='CONSULTAS')
plt.title('APGAR5 Médio por Consultas')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(12, 13))
fig, axes = plt.subplots(2, 2)


sns.countplot(ax=axes[0,0], x='SEXO', data=base_dados)
sns.countplot(ax=axes[0,1], x='GESTACAO', data=base_dados)
sns.countplot(ax=axes[1,0], x='ESTCIVMAE', data=base_dados)
sns.countplot(ax=axes[1,1], x='CONSULTAS', data=base_dados)

axes[0, 1].set_xticklabels(axes[0, 1].get_xticklabels(), rotation=45, ha='right')
plt.tight_layout()
```
